```python
## Loading Libraries
import pandas
import numpy
```

```python
## loading Data
gap = pandas.read_csv('gapminder.csv')
print(len(gap))
print(len(gap.columns))

## converting variable names to lower case
gap.columns=map(str.lower, gap.columns)
```

```python
## Creating a categorical variable
gap['incomeperperson_cat']=pandas.qcut(gap.incomeperperson, 4, 
    labels=["1st quartile","2nd quartile","3rd quartile","4th quartile"])
gap.head()
```

```python
print("income per person quartiles")
category1 = gap["incomeperperson_cat"].value_counts(sort=False)
print(category1)

print("\n")

print("income per person percentages")
category2 = gap["incomeperperson_cat"].value_counts(sort=False,normalize=True)
print(category2)
```

```python
print("income per person quartiles-second method")
category3 = gap.groupby('incomeperperson_cat').size()
print(category3)

print("\n")

print("income per person percentages-second method")
category4 = gap.groupby('incomeperperson_cat').size()*100/len(gap)
print(category4)
```

```python
## Creating a quantitative varibale
gap['incomeperperson'] = pandas.to_numeric(gap['incomeperperson'],errors='coerce')
quantity1 = gap["incomeperperson"].value_counts(sort=False)
print(quantity1)
```

```python
quantity2 = gap.groupby('incomeperperson').size()
print(quantity2)
```

```python
## Subsetting Data
gap['suicideper100th'] = pandas.to_numeric(gap['suicideper100th'],errors='coerce')

gap_refined = gap[(gap['suicideper100th']>0) & (gap['incomeperperson']>0) ]
gap_refined[['country','incomeperperson','suicideper100th']].head()
```
