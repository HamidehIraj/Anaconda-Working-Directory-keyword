```python
## Loading Libraries
import pandas
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
## Loading Data
gap = pandas.read_csv('gapminder.csv')
print(len(gap))
print(len(gap.columns))

## converting variable names to lower case
gap.columns=map(str.lower, gap.columns)
```

```python
## Preparing Data
gap['alcconsumption']   = pandas.to_numeric(gap['alcconsumption'],  errors='coerce')
gap['incomeperperson']  = pandas.to_numeric(gap['incomeperperson'], errors='coerce')

gap = gap[(gap['alcconsumption']>0) & (gap['incomeperperson']>0) ]

print(gap.shape)
```

```python
print("Alcohol Consumption Summary")
print(gap.alcconsumption.describe())

#Univariate histogram for quantitative variable:
seaborn.distplot(gap['alcconsumption'].dropna(), kde=False);
plt.title('The Frequency of Alcohol Consumption')
```

```python
print('Income per Person Summary')
print(gap.incomeperperson.describe())

#Univariate histogram for quantitative variable:
seaborn.distplot(gap['incomeperperson'].dropna(), kde=False);
plt.title('The Frequency of Income per Person')
```

```python
import warnings
warnings.filterwarnings("ignore")

#basic scatterplot:  Q->Q
scat1 = seaborn.regplot(x="incomeperperson", y="alcconsumption", fit_reg=False, data=gap)
plt.xlabel('Income per Person')
plt.ylabel('Alcohol Consumption')
plt.title('Scatterplot for the Association Between Alcohol Consumption and Income per Person')
```
