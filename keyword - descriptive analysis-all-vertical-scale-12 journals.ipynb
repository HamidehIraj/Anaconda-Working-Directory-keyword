# Preparing Data

```r
## Finding the difference in columns
diff <- setdiff(names(comp),names(evaluation ))
print(diff)
 
## creating columns
for (col_name in diff)
    
    ## Q1
    {
     if (!(col_name %in% names(internet)))
    {internet[,col_name] <- NA}
    
     if (!(col_name %in% names(teaching)))
    {teaching[,col_name] <- NA}
##------------------------------------------
    ## Q2
    
    if (!(col_name %in% names(rev)))
    {rev[,col_name] <- NA}
    
    if (!(col_name %in% names(teacher)))
    {teacher[,col_name] <- NA}
        
    if (!(col_name %in% names(techinhigh)))
    {techinhigh[,col_name] <- NA}
##---------------------------------------------
    ## Q3
    
    if (!(col_name %in% names(history)))
    {history[,col_name] <- NA}
    
        if (!(col_name %in% names(continue)))
    {continue[,col_name] <- NA}
##---------------------------------------------
    ## Q4
    
    if (!(col_name %in% names(compin)))
   { compin[,col_name] <- NA}
    
    if (!(col_name %in% names(evaluation)))
   { evaluation[,col_name] <- NA}
    
    if (!(col_name %in% names(pedagogies)))
   { pedagogies[,col_name] <- NA}
    }

## reordering  columns according to comp columns
## Q1
internet    <- internet   [,names(comp)]
teaching    <- teaching   [,names(comp)]
##---------------------------------------
## Q2
rev         <- rev        [,names(comp)]
teacher     <- teacher    [,names(comp)]
techinhigh  <- techinhigh [,names(comp)]
##---------------------------------------
## Q3
pers        <- pers       [,names(comp)]
history     <- history    [,names(comp)]
continue    <- continue   [,names(comp)]
##---------------------------------------
## Q4
compin      <- compin    [,names(comp)]
evaluation  <- evaluation[,names(comp)]
pedagogies  <- pedagogies[,names(comp)]



## Combining the datasets
data <- as.data.frame(rbind(comp,internet,teaching,        ##Q1
                            rev,teacher,techinhigh,        ##Q2
                            pers,history,continue,         ##Q3
                            compin,evaluation, pedagogies  ##Q4
                            
                           ),na.strings=c(""))


## Checking the number of rows
## Q1
dim(comp)
dim(internet)
dim(teaching)
## Q2
dim(rev)
dim(teacher)
dim(techinhigh)
## Q3
dim(pers)
dim(history)
dim(continue)
## Q4
dim(compin)
dim(evaluation)
dim(pedagogies)
print("----------------------------------------------------------")
nrow(data)

nrow(comp) + nrow(internet) + nrow(teaching) + 
nrow(rev) +nrow(teacher) + nrow(techinhigh)+ 
nrow(pers) + nrow(history)+nrow(continue)+
nrow(compin)+nrow(evaluation)+nrow(pedagogies)== nrow(data)


## Selecting research articles
# data <- data[data$document.type=="article",]


```

# Correlation Matrix (including the target  variable)

```r
## correcting predictors
predictors <- c("no_authors","article_age","page.count",
                "abstract_length","title_length",
                "abstract_ratio"
               )

library(repr)
## options(repr.plot.width=3, repr.plot.height=3,repr.plot.res = 40)
options(repr.plot.width=8, repr.plot.height=8)

## put correlations on the upper panels,
## with size proportional to the correlations.
panel.cor <- function(x, y, digits = 2, prefix = "", cex.cor, ...)
{
    usr <- par("usr"); on.exit(par(usr))
    par(usr = c(0, 1, 0, 1))
    # r <- abs(cor(x, y))

    r <- cor(x, y)
    txt <- format(c(r, 0.123456789), digits = digits)[1]
    txt <- paste0(prefix, txt)
    if(missing(cex.cor)) cex.cor <- 0.8/strwidth(txt)
    
    # removing the propotional property    
    # text(0.5, 0.5, txt, cex = cex.cor * r)
    text(0.5, 0.5, txt)
   
}
    
pairs(data[,c(predictors,target)],lower.panel = panel.smooth, upper.panel = panel.cor)
```
