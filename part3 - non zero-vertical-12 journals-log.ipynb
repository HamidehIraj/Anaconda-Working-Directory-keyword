# Part 3: Linear Regression in citation >0

```r
require(car,quietly=TRUE)
pred_numeric <- c("abstract_length" ,"page.count","title_length", "abstract_ratio" ,"no_references","no_authors"
                  ,"numkeys", "cited.by", "article_age","weight_ratio")
pred_cat     <- c("quartile")
pred_data     <- data[,c(pred_numeric,pred_cat)]
nrow(data)

linear_reg   <- lm(cited.by^ -0.05 ~ abstract_length + log(page.count) + title_length + 
    abstract_ratio + log(no_references) + log(no_authors) + numkeys + article_age + 
    weight_ratio + quartile
                       , data=pred_data)
summary(linear_reg)

# or
# linear_reg2   <- lm(bcPower(cited.by , -0.05)  ~ ., data=pred_data)
# summary(linear_reg2)
#round(linear_reg$coeff,3)
```
